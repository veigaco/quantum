```python
import pandas as pd
from datetime import time
from time import sleep
```

```python
SUMMARY_URL = 'https://finance.yahoo.com/quote/{}?p={}'
SECTION_URL = 'https://finance.yahoo.com/quote/{}{}?p={}'
ticker = 'SBUX'
# https://finance.yahoo.com/quote/SBUX?p=SBUX -> Summary 
# https://finance.yahoo.com/quote/SBUX/key-statistics?p=SBUX -> Statistics
# https://finance.yahoo.com/quote/SBUX/profile?p=SBUX -> Profile
# https://finance.yahoo.com/quote/SBUX/financials?p=SBUX -> Income Statement
# https://finance.yahoo.com/quote/SBUX/balance-sheet?p=SBUX -> Balance Sheet
# https://finance.yahoo.com/quote/SBUX/cash-flow?p=SBUX -> Cash Flow
# https://finance.yahoo.com/quote/SBUX/holders?p=SBUX -> Holders
# https://finance.yahoo.com/quote/SBUX/analysis?p=SBUX -> Analysts

sections = ['/key-statistics', '/profile', '/financials', '/balance-sheet', '/cash-flow', '/holders', '/analysis']
urls = [SUMMARY_URL.format(ticker, ticker)]
urls.extend([SECTION_URL.format(ticker, s, ticker) for s in sections])
```

```python
for u in urls:
    print("Downloading information for:", u)
    tables = profile = pd.read_html(u)
    for t in tables:
        print(t)
    sleep(1)
```
