```python
import pandas as pd
from datetime import datetime,date,timedelta
import time
import pandas as pd
from sqlalchemy.orm import sessionmaker#,relationship, backref
from sqlalchemy import create_engine,select
from Schema import *
from mvo_utils import *
```

```python
## cargar datos de bloomberg
data_bloomberg=pd.read_csv("./data_FRED_Bloomberg_inicial/Bloomberg.csv",index_col=[0])
data_bloomberg["instrument_id"]=data_bloomberg["ticker"].apply(get_id_instrument)
data_bloomberg["date_"]=pd.to_datetime(data_bloomberg["date_"])
#data_bloomberg.to_sql(con=engine, name='time_series', if_exists='append',index=False) 
data_bloomberg.tail()
```

```python
data_bloomberg.to_sql(con=engine, name='time_series', if_exists='append',index=False) 
```

```python
# update instrument table
l=data_bloomberg.ticker.unique().tolist()
for i in l:
    aux=data_bloomberg[data_bloomberg.ticker==i]
    df_instrument = pd.DataFrame(columns=["refreshed_at","newest_available_date","oldest_available_date","start_date","end_date","frequency"])
    df_instrument.loc[0] = [datetime.fromtimestamp(time.time()).strftime('%Y-%m-%dT%H:%M:%S.%fZ'),max(aux.date_).strftime('%Y-%m-%d'),min(aux.date_).strftime('%Y-%m-%d'),min(aux.date_).strftime('%Y-%m-%d'),max(aux.date_).strftime('%Y-%m-%d'),""]
    df_instrument
    aux.columns=["date_","value","ticker","category","id"]
    row=aux.loc[0]
    #print(row)
    update_data_vendor("Bloomberg","https://www.bloomberg.com/")
    update_instrument_db(df_instrument,row)
```
